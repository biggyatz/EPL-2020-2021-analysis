```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample player performance data (you can replace this with your actual dataset)
player_data = {
    'Player': ['Player A', 'Player B', 'Player C', 'Player D', 'Player E'],
    'Goals': [10, 8, 5, 12, 6],
    'Assists': [5, 6, 3, 7, 4],
    'Shots_on_target': [30, 25, 20, 35, 22],
    'Pass_completion_rate': [80, 85, 78, 87, 82],
    'Interceptions': [15, 10, 12, 18, 8]
}

# Create a DataFrame from the player data
df = pd.DataFrame(player_data)

# Calculate some player performance metrics
df['Total_Shots'] = df['Shots_on_target'] / 0.6  # Assuming 60% of shots are on target
df['Goal_Conversion_Rate'] = (df['Goals'] / df['Total_Shots']) * 100
df['Assist_per_Goal'] = df['Assists'] / df['Goals']

# Visualize goal conversion rate
plt.figure(figsize=(10, 6))
plt.bar(df['Player'], df['Goal_Conversion_Rate'], color='skyblue')
plt.title('Goal Conversion Rate of Players')
plt.xlabel('Player')
plt.ylabel('Goal Conversion Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Visualize assists per goal
plt.figure(figsize=(10, 6))
plt.bar(df['Player'], df['Assist_per_Goal'], color='lightgreen')
plt.title('Assists per Goal of Players')
plt.xlabel('Player')
plt.ylabel('Assists per Goal')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample player performance data (you can replace this with your actual dataset)
player_data = {
    'Player': ['Player A', 'Player B', 'Player C', 'Player D', 'Player E'],
    'Goals': [10, 8, 5, 12, 6],
    'Assists': [5, 6, 3, 7, 4],
    'Shots_on_target': [30, 25, 20, 35, 22],
    'Pass_completion_rate': [80, 85, 78, 87, 82],
    'Interceptions': [15, 10, 12, 18, 8]
}

# Create a DataFrame from the player data
df = pd.DataFrame(player_data)

# Calculate some player performance metrics
df['Total_Shots'] = df['Shots_on_target'] / 0.6  # Assuming 60% of shots are on target
df['Goal_Conversion_Rate'] = (df['Goals'] / df['Total_Shots']) * 100
df['Assist_per_Goal'] = df['Assists'] / df['Goals']

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plotting the heatmap with player names
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, 
            xticklabels=corr_matrix.columns, yticklabels=corr_matrix.columns)
plt.title('Correlation Matrix of Player Performance Metrics')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ggplot import *

# Sample shot statistics data
shot_stats = {
    'Stat': ['Total Shots', 'Shots on Target', 'Shots off Target', 'Goals'],
    'Value': [50, 30, 20, 10]
}

# Create a DataFrame from the shot statistics data
df_shot_stats = pd.DataFrame(shot_stats)

# Define the dimensions of the goal post
goal_width = 7.32  # in meters
goal_height = 2.44  # in meters

# Define the coordinates for plotting shot statistics
x_coordinates = [2, 6, 4, 4]  # x-coordinates for Total Shots, Shots on Target, Shots off Target, and Goals
y_coordinates = [1, 1, 1, 2]  # y-coordinates for Total Shots, Shots on Target, Shots off Target, and Goals

# Create a custom plot
plt.figure(figsize=(10, 6))

# Plot the goal post
plt.fill_betweenx([0, goal_height], 0, goal_width, color='black', alpha=0.7)

# Plot shot statistics
for i, row in df_shot_stats.iterrows():
    plt.scatter(x_coordinates[i], y_coordinates[i], s=row['Value']*10, label=row['Stat'])

# Set axis limits and labels
plt.xlim(-1, 8)
plt.ylim(0, 3)
plt.xticks([])
plt.yticks([])
plt.title('Football Shot Statistics Visualization')

# Add legend
plt.legend(loc='upper right')

# Add annotations for shot statistics
for i, row in df_shot_stats.iterrows():
    plt.annotate(f'{row["Stat"]}: {row["Value"]}', (x_coordinates[i], y_coordinates[i]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=10, color='black')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample shot statistics data
shot_stats = {
    'Total Shots': 50,
    'Shots on Target': 30,
    'Shots off Target': 20,
    'Goals': 10
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Define the dimensions of the goal post
goal_width = 7.32  # in meters
goal_height = 2.44  # in meters

# Plot the goal post
ax.add_patch(plt.Rectangle((0, 0), goal_width, goal_height, color='black', alpha=0.7))

# Plot shots on target
ax.scatter(2, 1, s=shot_stats['Shots on Target']*10, color='green', alpha=0.8, label='Shots on Target')
# Plot shots off target
ax.scatter(6, 1, s=shot_stats['Shots off Target']*10, color='red', alpha=0.5, label='Shots off Target')
# Plot goals
ax.scatter(4, 1, s=shot_stats['Goals']*20, color='blue', alpha=0.8, label='Goals')
# Plot total shots
ax.scatter(4, 2, s=shot_stats['Total Shots']*5, color='orange', alpha=0.5, label='Total Shots')

# Set axis limits and labels
ax.set_xlim(-1, 8)
ax.set_ylim(0, 3)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Football Shot Statistics Visualization')

# Add legend
ax.legend(loc='upper right')

# Add annotations for shot statistics
for stat, (x, y) in zip(shot_stats.keys(), [(2, 1), (6, 1), (4, 1), (4, 2)]):
    ax.annotate(f'{stat}: {shot_stats[stat]}', (x, y), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=10, color='black')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Sample shot statistics data
shot_stats = {
    'Stat': ['Total Shots', 'Shots on Target', 'Shots off Target', 'Goals'],
    'Value': [50, 30, 20, 10]
}

# Define the dimensions of the goal post
goal_width = 7.32  # in meters
goal_height = 2.44  # in meters

# Define the coordinates for plotting shot statistics
x_coordinates = [2, 6, 4, 4]  # x-coordinates for Total Shots, Shots on Target, Shots off Target, and Goals
y_coordinates = [1, 1, 1, 2]  # y-coordinates for Total Shots, Shots on Target, Shots off Target, and Goals

# Create a custom plot
fig, ax = plt.subplots(figsize=(10, 6))

# Load background image of football field
img = plt.imread('football_field.png')
ax.imshow(img, extent=[0, goal_width, 0, goal_height], alpha=0.5)

# Plot shot statistics
for x, y, stat, value in zip(x_coordinates, y_coordinates, shot_stats['Stat'], shot_stats['Value']):
    ax.scatter(x, y, s=value*10, label=stat, color='blue', alpha=0.7, edgecolors='black', linewidth=1)

# Set axis limits and labels
ax.set_xlim(0, goal_width)
ax.set_ylim(0, goal_height)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Football Shot Statistics Visualization')

# Add legend
legend = ax.legend(loc='upper right', fontsize='large')
for handle in legend.legendHandles:
    handle.set_sizes([100])

# Add annotations for shot statistics
for x, y, stat, value in zip(x_coordinates, y_coordinates, shot_stats['Stat'], shot_stats['Value']):
    ax.text(x, y, f'{stat}: {value}', ha='center', va='center', fontsize='large', color='black')

plt.show()
```
